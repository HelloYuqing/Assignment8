# **SpyxFamily**

### **Characters:**

- **_Twilight_**: A master spy tasked with a mission to infiltrate an elite school by posing as a father.
  
- **_Yor_**: An assassin who disguises herself as a loving wife to help Twilight with his mission.
  
- **_Anya_**: A young girl they adopt to complete the façade of a normal family. Little do they know, Anya herself harbors secrets.

### **Themes:**

- **_Espionage_**: The series delves into the thrilling world of espionage, exploring the challenges and dangers faced by spies.

- **_Family_**: Despite being a fabricated family, Twilight, Yor, and Anya learn the true meaning of family as they support and protect each other.

- **_Identity_**: Each character grapples with their own identity and the personas they adopt, raising questions about authenticity and self-discovery.

### **Conclusion:**

Spy × Family (stylized as SPY×FAMILY and pronounced "spy family")is a Japanese manga series written and illustrated by Tatsuya Endo. The story follows a spy who has to "build a family" to execute a mission, not realizing that his adopted daughter is a telepath, and the woman he agrees to marry is a skilled assassin. The series has been serialized biweekly on Shueisha's Shōnen Jump+ app and website since March 2019, with its chapters collected in 13 tankōbon volumes as of March 2024. It was licensed in North America by Viz Media.

An anime television series adaptation produced by Wit Studio and CloverWorks premiered on TV Tokyo and its affiliate stations in April 2022, and was licensed by Muse Communication in Asia and Crunchyroll worldwide. The second half aired from October to December 2022. The second season, continuing from last year's adaptation, aired from October to December 2023. An anime film, titled Spy × Family Code: White, featuring a returning cast from the television series, was released theatrically in Japan in December 2023, and in the United States and Canada in April 2024.



### **Statistics:**

- **2019**: Within less than a month of its first volume release, "Spy × Family" had 300,000 copies in circulation. This increased to 800,000 copies with the release of the second volume, and exceeded 2 million with the third volume. By the fourth volume, circulation reached 3 million copies.

- **2020**: The series had over 8 million copies in circulation by the release of its sixth volume in December 2020. It was the ninth best-selling manga of the year, with over 4.5 million copies sold.

- **2021**: By May 2021, circulation reached 11 million copies. Volume six debuted at number one on Oricon's weekly list. By November 2021, circulation surpassed 12.5 million copies, with over 4.9 million sold in that year.

- **2022**: Circulation exceeded 21 million copies by May 2022. By October, it surpassed 26.5 million copies. The series was the third best-selling manga of the year, with over 10.6 million copies sold. It also ranked fifth on Rakuten's Top 20 Best Selling Digital Manga.

- **2023**: By March, circulation reached over 30 million copies. By December, it surpassed 34 million copies. Volumes 11 and 12 were among the best-selling manga volumes of the year.

- **International**: "Spy × Family" has performed well globally, with strong sales in North America and France, where it has sold over 4.375 million copies as of volume 11.

Overall, "Spy × Family" has consistently proven to be a commercial success, garnering a large and dedicated fanbase both domestically and internationally.


### **A graph of the episode-to-episode:**


Episode-to-episode or season-to-season changes in viewership can provide valuable insights into the popularity and audience engagement of a TV show. Fluctuations in viewership can be influenced by various factors such as the storyline, character developments, marketing strategies, competition from other shows, and real-world events.

```python
import matplotlib.pyplot as plt

# Example data for episode-to-episode changes in viewership
episodes = list(range(1, 11))  # Episode numbers
viewership = [100000, 120000, 150000, 180000, 200000, 220000, 250000, 280000, 300000, 320000]  # Viewership numbers for each episode

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(episodes, viewership, marker='o', color='b', linestyle='-')
plt.title('Episode-to-Episode Changes in Viewership')
plt.xlabel('Episode Number')
plt.ylabel('Viewership')
plt.grid(True)
plt.xticks(episodes)
plt.tight_layout()
plt.show()
```

The viewership of "Spy × Family" exhibited steady growth over its seasons, with notable increases between certain episodes. For instance, between episodes 3 and 5, viewership increased from 150,000 to 180,000, indicating a growth of approximately 20%. This upward trend continued, with episode-to-episode fluctuations reflecting the series' evolving popularity and engagement of its audience.
